```python
import tkinter as tk
from tkinter import messagebox
import datetime

class RestaurantManagementSystem:
    def __init__(self, root):
        self.root = root
        self.root.title("Restaurant Management System")
        self.root.geometry("800x600")

        # Colors
        self.bg_color = "#00aaff"  # Aqua Blue
        self.text_color = "white"
        self.send_bill_color = "#0077cc"  # Ocean Blue

        # Variables
        self.lassi = tk.IntVar()
        self.coffee = tk.IntVar()
        self.tea = tk.IntVar()
        self.juice = tk.IntVar()
        self.shakes = tk.IntVar()
        self.milk = tk.IntVar()
        self.shikanji = tk.IntVar()
        self.redbull = tk.IntVar()

        self.roti = tk.IntVar()
        self.dal_makhni = tk.IntVar()
        self.mutter_paneer = tk.IntVar()
        self.paratha = tk.IntVar()
        self.mix_veg = tk.IntVar()
        self.omelette = tk.IntVar()
        self.veg_biryani = tk.IntVar()
        self.rice = tk.IntVar()

        self.customer_name = tk.StringVar()
        self.contact_number = tk.StringVar()
        self.email = tk.StringVar()

        self.cost_of_drinks = tk.StringVar()
        self.cost_of_foods = tk.StringVar()
        self.service_charge = tk.StringVar()
        self.paid_tax = tk.StringVar()
        self.sub_total = tk.StringVar()
        self.total_cost = tk.StringVar()

        # Title
        title_frame = tk.Frame(self.root, bd=5, relief=tk.RIDGE, pady=10, bg=self.bg_color)
        title_frame.pack(side=tk.TOP, fill=tk.X)
        title_label = tk.Label(title_frame, text="Restaurant Management System", font=("Arial", 15, "bold"), bg=self.bg_color, fg=self.text_color)
        title_label.pack()

        # Customer Info Frame
        customer_frame = tk.LabelFrame(self.root, text="Customer Information", font=("Arial", 10, "bold"), bd=5, relief=tk.RIDGE, padx=10, bg=self.bg_color, fg=self.text_color)
        customer_frame.pack(side=tk.TOP, fill=tk.X, padx=10, pady=5)

        self.create_label_entry(customer_frame, "Customer Name", self.customer_name, 0)
        self.create_label_entry(customer_frame, "Contact Number", self.contact_number, 1)
        self.create_label_entry(customer_frame, "Email", self.email, 2)

        # Drinks Frame
        drinks_frame = tk.LabelFrame(self.root, text="Drinks", font=("Arial", 10, "bold"), bd=5, relief=tk.RIDGE, padx=10, bg=self.bg_color, fg=self.text_color)
        drinks_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        self.create_checkbutton(drinks_frame, "Lassi", self.lassi, 0)
        self.create_checkbutton(drinks_frame, "Coffee", self.coffee, 1)
        self.create_checkbutton(drinks_frame, "Tea", self.tea, 2)
        self.create_checkbutton(drinks_frame, "Juice", self.juice, 3)
        self.create_checkbutton(drinks_frame, "Shakes", self.shakes, 4)
        self.create_checkbutton(drinks_frame, "Milk", self.milk, 5)
        self.create_checkbutton(drinks_frame, "Shikanji", self.shikanji, 6)
        self.create_checkbutton(drinks_frame, "Redbull", self.redbull, 7)

        # Foods Frame
        foods_frame = tk.LabelFrame(self.root, text="Foods", font=("Arial", 10, "bold"), bd=5, relief=tk.RIDGE, padx=10, bg=self.bg_color, fg=self.text_color)
        foods_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        self.create_checkbutton(foods_frame, "Roti", self.roti, 0)
        self.create_checkbutton(foods_frame, "Dal Makhni", self.dal_makhni, 1)
        self.create_checkbutton(foods_frame, "Mutter Paneer", self.mutter_paneer, 2)
        self.create_checkbutton(foods_frame, "Paratha", self.paratha, 3)
        self.create_checkbutton(foods_frame, "Mix Veg", self.mix_veg, 4)
        self.create_checkbutton(foods_frame, "Omelette", self.omelette, 5)
        self.create_checkbutton(foods_frame, "Veg Biryani", self.veg_biryani, 6)
        self.create_checkbutton(foods_frame, "Rice", self.rice, 7)

        # Cost Frame
        cost_frame = tk.Frame(self.root, bd=5, relief=tk.RIDGE, padx=10, pady=10, bg=self.bg_color)
        cost_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        self.create_label_entry(cost_frame, "Cost of Drinks", self.cost_of_drinks, 0)
        self.create_label_entry(cost_frame, "Cost of Foods", self.cost_of_foods, 1)
        self.create_label_entry(cost_frame, "Service Charge", self.service_charge, 2)
        self.create_label_entry(cost_frame, "Paid Tax", self.paid_tax, 3)
        self.create_label_entry(cost_frame, "Sub Total", self.sub_total, 4)
        self.create_label_entry(cost_frame, "Total Cost", self.total_cost, 5)

        # Buttons Frame
        buttons_frame = tk.Frame(self.root, bd=5, relief=tk.RIDGE, bg=self.bg_color)
        buttons_frame.pack(side=tk.BOTTOM, fill=tk.BOTH, expand=True)

        total_button = tk.Button(buttons_frame, text="Calculate Total", command=self.calculate_total, font=("Arial", 10, "bold"), bd=3, bg=self.bg_color, fg=self.text_color)
        total_button.grid(row=0, column=0, padx=10)
        generate_button = tk.Button(buttons_frame, text="Generate Bill", command=self.generate_bill, font=("Arial", 10, "bold"), bd=3, bg=self.bg_color, fg=self.text_color)
        generate_button.grid(row=0, column=1, padx=10)
        reset_button = tk.Button(buttons_frame, text="Reset", command=self.reset, font=("Arial", 10, "bold"), bd=3, bg=self.bg_color, fg=self.text_color)
        reset_button.grid(row=0, column=2, padx=10)
        exit_button = tk.Button(buttons_frame, text="Exit", command=self.exit, font=("Arial", 10, "bold"), bd=3, bg=self.bg_color, fg=self.text_color)
        exit_button.grid(row=0, column=3, padx=10)

        # Calculator Frame
        calculator_frame = tk.Frame(self.root, bd=5, relief=tk.RIDGE, bg=self.bg_color)
        calculator_frame.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

        self.calculator_display = tk.Entry(calculator_frame, font=("Arial", 12, "bold"), bd=3, insertwidth=4, width=14, justify='right')
        self.calculator_display.grid(row=0, column=0, columnspan=4)

        calculator_buttons = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('+', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('-', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('*', 3, 3),
            ('0', 4, 0), ('C', 4, 1), ('=', 4, 2), ('/', 4, 3),
        ]

        for (text, row, column) in calculator_buttons:
            self.create_calculator_button(calculator_frame, text, row, column)

        # Receipt Frame
        receipt_frame = tk.Frame(self.root, bd=5, relief=tk.RIDGE, bg=self.bg_color)
        receipt_frame.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

        self.receipt_text = tk.Text(receipt_frame, width=34, height=12, font=("Arial", 10, "bold"), bd=3, bg=self.bg_color, fg=self.text_color)
        self.receipt_text.grid(row=0, column=0)

    def create_checkbutton(self, frame, text, variable, row):
        checkbutton = tk.Checkbutton(frame, text=text, variable=variable, font=("Arial", 10), bg=self.bg_color, fg=self.text_color, onvalue=1, offvalue=0)
        checkbutton.grid(row=row, column=0, sticky="w")

    def create_label_entry(self, frame, text, variable, row):
        label = tk.Label(frame, text=text, font=("Arial", 10), bg=self.bg_color, fg=self.text_color)
        label.grid(row=row, column=0, sticky="w", padx=10, pady=5)
        entry = tk.Entry(frame, textvariable=variable, font=("Arial", 10), bd=3)
        entry.grid(row=row, column=1, padx=10, pady=5)

    def create_calculator_button(self, frame, text, row, column):
        button = tk.Button(frame, text=text, padx=16, pady=1, font=("Arial", 10, "bold"), bd=4, bg=self.bg_color, fg=self.text_color)
        button.grid(row=row, column=column)
        button.config(command=lambda: self.on_calculator_button_click(text))

    def on_calculator_button_click(self, text):
        current_text = self.calculator_display.get()
        if text == "C":
            self.calculator_display.delete(0, tk.END)
        elif text == "=":
            try:
                result = eval(current_text)
                self.calculator_display.delete(0, tk.END)
                self.calculator_display.insert(tk.END, str(result))
            except:
                self.calculator_display.delete(0, tk.END)
                self.calculator_display.insert(tk.END, "Error")
        else:
            self.calculator_display.insert(tk.END, text)

    def calculate_total(self):
        try:
            cost_drinks = (
                self.lassi.get() * 30 +
                self.coffee.get() * 40 +
                self.tea.get() * 20 +
                self.juice.get() * 50 +
                self.shakes.get() * 60 +
                self.milk.get() * 25 +
                self.shikanji.get() * 20 +
                self.redbull.get() * 100
            )
            cost_foods = (
                self.roti.get() * 10 +
                self.dal_makhni.get() * 50 +
                self.mutter_paneer.get() * 70 +
                self.paratha.get() * 30 +
                self.mix_veg.get() * 60 +
                self.omelette.get() * 20 +
                self.veg_biryani.get() * 80 +
                self.rice.get() * 40
            )
            service_charge = 50
            tax = (cost_drinks + cost_foods + service_charge) * 0.05
            sub_total = cost_drinks + cost_foods + service_charge
            total_cost = sub_total + tax

            self.cost_of_drinks.set(f"Rs. {cost_drinks:.2f}")
            self.cost_of_foods.set(f"Rs. {cost_foods:.2f}")
            self.service_charge.set(f"Rs. {service_charge:.2f}")
            self.paid_tax.set(f"Rs. {tax:.2f}")
            self.sub_total.set(f"Rs. {sub_total:.2f}")
            self.total_cost.set(f"Rs. {total_cost:.2f}")

            self.receipt_text.delete("1.0", tk.END)
            self.receipt_text.insert(tk.END, f"Cost of Drinks: Rs. {cost_drinks:.2f}\n")
            self.receipt_text.insert(tk.END, f"Cost of Foods: Rs. {cost_foods:.2f}\n")
            self.receipt_text.insert(tk.END, f"Service Charge: Rs. {service_charge:.2f}\n")
            self.receipt_text.insert(tk.END, f"Paid Tax: Rs. {tax:.2f}\n")
            self.receipt_text.insert(tk.END, f"Sub Total: Rs. {sub_total:.2f}\n")
            self.receipt_text.insert(tk.END, f"Total Cost: Rs. {total_cost:.2f}\n")

        except Exception as e:
            messagebox.showerror("Error", str(e))

    def generate_bill(self):
        try:
            self.calculate_total()
            bill_no = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
            bill_date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            self.receipt_text.insert(tk.END, f"\nBill No: {bill_no}")
            self.receipt_text.insert(tk.END, f"\nDate: {bill_date}")
            self.receipt_text.insert(tk.END, f"\nCustomer: {self.customer_name.get()}")
            self.receipt_text.insert(tk.END, f"\nContact: {self.contact_number.get()}")
            self.receipt_text.insert(tk.END, f"\nEmail: {self.email.get()}\n")

            self.receipt_text.insert(tk.END, f"\nItems\t\tQuantity\tAmount\n")
            self.receipt_text.insert(tk.END, f"-----------------------------------------\n")
            self.add_receipt_item("Lassi", self.lassi, 30)
            self.add_receipt_item("Coffee", self.coffee, 40)
            self.add_receipt_item("Tea", self.tea, 20)
            self.add_receipt_item("Juice", self.juice, 50)
            self.add_receipt_item("Shakes", self.shakes, 60)
            self.add_receipt_item("Milk", self.milk, 25)
            self.add_receipt_item("Shikanji", self.shikanji, 20)
            self.add_receipt_item("Redbull", self.redbull, 100)

            self.add_receipt_item("Roti", self.roti, 10)
            self.add_receipt_item("Dal Makhni", self.dal_makhni, 50)
            self.add_receipt_item("Mutter Paneer", self.mutter_paneer, 70)
            self.add_receipt_item("Paratha", self.paratha, 30)
            self.add_receipt_item("Mix Veg", self.mix_veg, 60)
            self.add_receipt_item("Omelette", self.omelette, 20)
            self.add_receipt_item("Veg Biryani", self.veg_biryani, 80)
            self.add_receipt_item("Rice", self.rice, 40)

            # Show Send Bill Window
            self.show_send_bill_window()

        except Exception as e:
            messagebox.showerror("Error", str(e))

    def add_receipt_item(self, item_name, item_var, price):
        if item_var.get() > 0:
            self.receipt_text.insert(tk.END, f"{item_name}\t\t{item_var.get()}\t\t{item_var.get() * price}\n")

    def show_send_bill_window(self):
        send_bill_window = tk.Toplevel(self.root)
        send_bill_window.title("Send Bill")
        send_bill_window.geometry("400x300")
        send_bill_window.configure(bg=self.send_bill_color)

        # Bill Details
        bill_details_frame = tk.Frame(send_bill_window, bg=self.send_bill_color)
        bill_details_frame.pack(pady=10, padx=10, fill=tk.BOTH, expand=True)

        tk.Label(bill_details_frame, text=f"Customer Name: {self.customer_name.get()}", font=("Arial", 12), bg=self.send_bill_color, fg="white").pack(anchor="w")
        tk.Label(bill_details_frame, text=f"Contact Number: {self.contact_number.get()}", font=("Arial", 12), bg=self.send_bill_color, fg="white").pack(anchor="w")
        tk.Label(bill_details_frame, text=f"Email: {self.email.get()}", font=("Arial", 12), bg=self.send_bill_color, fg="white").pack(anchor="w")

        tk.Label(bill_details_frame, text="\nBill Details:", font=("Arial", 12, "bold"), bg=self.send_bill_color, fg="white").pack(anchor="w")
        tk.Text(bill_details_frame, height=10, width=50, font=("Arial", 10), bg="white", fg="black", state=tk.NORMAL).pack(pady=5)
        tk.Label(bill_details_frame, text="--- End of Bill ---", font=("Arial", 10), bg=self.send_bill_color, fg="white").pack(anchor="w")

        # Send Bill Button
        send_button = tk.Button(bill_details_frame, text="Send Bill", font=("Arial", 12, "bold"), bg="green", fg="white", command=self.send_bill)
        send_button.pack(pady=10)

    def send_bill(self):
        # Logic for sending the bill
        messagebox.showinfo("Send Bill", "Bill has been sent successfully!")

    def reset(self):
        self.lassi.set(0)
        self.coffee.set(0)
        self.tea.set(0)
        self.juice.set(0)
        self.shakes.set(0)
        self.milk.set(0)
        self.shikanji.set(0)
        self.redbull.set(0)
        self.roti.set(0)
        self.dal_makhni.set(0)
        self.mutter_paneer.set(0)
        self.paratha.set(0)
        self.mix_veg.set(0)
        self.omelette.set(0)
        self.veg_biryani.set(0)
        self.rice.set(0)
        self.customer_name.set("")
        self.contact_number.set("")
        self.email.set("")
        self.cost_of_drinks.set("")
        self.cost_of_foods.set("")
        self.service_charge.set("")
        self.paid_tax.set("")
        self.sub_total.set("")
        self.total_cost.set("")
        self.receipt_text.delete("1.0", tk.END)

    def exit(self):
        self.root.destroy()

if __name__== "__main__":
    root = tk.Tk()
    app = RestaurantManagementSystem(root)
    root.mainloop()
```
